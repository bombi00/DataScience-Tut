## Matplotlib example workflow

```python
#1 import matplot
%matplotlib inline
import matplotlib.pyplot as plt

#2 prepare data
x = [1,2,3,4]
y = [2,33,24,5]

#3 Setup plot
fig, ax = plt.subplots(figsize=(10,10))

#4 Plot data
ax.plot(x, y)

#5 Customized

ax.set(title="Sample Plot",
       xlabel="x-axis",
       ylabel="y-axis")

#6 Save and Show
fig.savefig("images/sampleplot.png")
```

## Plotting from Pandas DataFrame

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
car_sales["Data sales"] = pd.date_range("1-1-2020", periods=len(car_sales))
```

```python
car_sales
```

```python
car_sales["Total sales"] = car_sales["Price"].cumsum()
car_sales
```

```python
car_sales.plot(x="Data sales", y="Total sales");
```

```python
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter", s=5)
mean_price = car_sales["Price"].mean()
plt.axhline(y = 7000, color="red", linestyle='dashdot')

# Mostra il plot
plt.show();
```

```python
#Another Dataset
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
heart_disease["age"].plot.hist();
```

```python
heart_disease.plot.hist(figsize=(10, 50), subplots=True, bins=20);
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
# Pyplot method
over_50.plot(kind='scatter', x="age", y="chol", c="target");
```

```python
# OO Method mixed

fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter', x="age", y="chol", c="target", ax=ax)
```

```python
## OO Method
fig, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"]);
ax.set(title="Heart disease and Cholesterol level"  ,xlabel="Age", ylabel="Cholesterol");
ax.legend(*scatter.legend_elements(), title="Target")
ax.axhline(over_50["chol"].mean(), linestyle="--", color="red");
```

```python
plt.style.use('ggplot')

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(10,10))
scatter1 = ax1.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"], cmap="winter");
ax1.set(title="Heart disease and Cholesterol level"  ,xlabel="Age", ylabel="Cholesterol");
ax1.set_xlim([50, 80])
ax1.legend(*scatter1.legend_elements(), title="Target")
ax1.axhline(over_50["chol"].mean(), linestyle="--", color="red");

scatter1 = ax2.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"], cmap="winter");
ax2.set(title="Heart disease and Max Heart Rate"  ,xlabel="Age", ylabel="Max Heart Rate");
ax2.set_xlim([50, 80])
ax2.set_ylim([60, 200])
ax2.legend(*scatter1.legend_elements(), title="Target")
ax2.axhline(over_50["thalach"].mean(), linestyle="--", color="red");

fig.suptitle("Heart disease Analysis", fontsize=16, fontweight="bold");
```
