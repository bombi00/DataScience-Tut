## Matplotlib example workflow

```python
#1 import matplot
%matplotlib inline
import matplotlib.pyplot as plt

#2 prepare data
x = [1,2,3,4]
y = [2,33,24,5]

#3 Setup plot
fig, ax = plt.subplots(figsize=(10,10))

#4 Plot data
ax.plot(x, y)

#5 Customized

ax.set(title="Sample Plot",
       xlabel="x-axis",
       ylabel="y-axis")

#6 Save and Show
fig.savefig("images/sampleplot.png")
```

# Making Plot from NumpyData

```python
x = np.linspace(1,10,100)
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x), s=5);
```

```python
#From dictionary
butter = {"Almond butter" : 10,
          "Peanut butter" : 8,
          "Coconut butter" : 15}

fig, ax = plt.subplots()
ax.bar(butter.keys(), butter.values());
ax.set(title="Creme spalmabili", ylabel="Costo in $");
```

```python
x = np.random.randn(1000) #Normal distribution
fig, ax = plt.subplots()
ax.hist(x);
```

```python
#Plotting more data
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5));

ax1.plot(x, x/2)
ax2.bar(butter.keys(), butter.values())
ax3.scatter(np.random.rand(100), np.random.rand(100), s=5)
ax4.hist(x);
```

## Plotting from Pandas DataFrame

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].replace("[$,.]", "", regex=True).astype(int).div(100)
car_sales
```

```python
car_sales["Data sales"] = pd.date_range("1-1-2020", periods=len(car_sales))
```

```python
car_sales
```

```python
car_sales["Total sales"] = car_sales["Price"].cumsum()
car_sales
```

```python
car_sales.plot(x="Data sales", y="Total sales");
```
